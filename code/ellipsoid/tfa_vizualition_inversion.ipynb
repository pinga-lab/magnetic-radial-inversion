```python
verts = []
for o in di['prisms']:
    top = []
    bottom = []
    for x, y in zip(o.x, o.y):
        top.append(np.array([x,y,o.z1]))
        bottom.append(np.array([x,y,o.z2]))
    verts.append(top)
    verts.append(bottom)
    for i in range(o.x.size-1):
        sides = []
        sides.append(np.array([o.x[i], o.y[i], o.z1]))
        sides.append(np.array([o.x[i+1], o.y[i+1], o.z1]))
        sides.append(np.array([o.x[i+1], o.y[i+1], o.z2]))
        sides.append(np.array([o.x[i], o.y[i], o.z2]))
        verts.append(sides)
    sides = []
    sides.append(np.array([o.x[-1], o.y[-1], o.z1]))
    sides.append(np.array([o.x[0], o.y[0], o.z1]))
    sides.append(np.array([o.x[0], o.y[0], o.z2]))
    sides.append(np.array([o.x[-1], o.y[-1], o.z2]))
    verts.append(sides)
```

```python
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')

rx, ry, rz = (300., 150., 25.)

# Set of all spherical angles:
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)

# Cartesian coordinates that correspond to the spherical angles:
# (this is the equation of an ellipsoid):
x = rx * np.outer(np.cos(u), np.sin(v))
y = ry * np.outer(np.sin(u), np.sin(v))
z = rz * np.outer(np.ones_like(u), np.cos(v)) + 150.

# Plot:
ax.plot_surface(x, y, z,  rstride=4, cstride=4, color='b', alpha=1)

# plot sides
ax.add_collection3d(Poly3DCollection(verts, alpha=0.1, 
 facecolor='r', linewidths=0.5, edgecolors='k'))

ax.set_xlim(-200., 200., 50)
ax.set_ylim(-200., 200., 50)
ax.set_zlim(300,50,50)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.savefig(mypath+'diorama_3d_estimate.png', dpi=300)
#plt.show()

img(filename=mypath+'diorama_3d_estimate.png', width=400, height=400)
```

```python
h = np.zeros(di['xp'].size) + 300  # layer depth
A_tf = msf.sm_tf_sphere(di['xp'], di['yp'], di['zp'], di['xp'], di['yp'], h, di['main_field'][0], di['main_field'][1], -40., di['inc_dec'][1])
```

```python
I = np.identity(di['xp'].size)
f0 = np.trace(np.dot(A_tf.T, A_tf))/di['xp'].size
u0 = 0.001
```

```python
vp = np.linalg.solve(np.dot(A_tf.T,A_tf) + u0*f0*I,np.dot(A_tf.T,di['residual']))
```

```python
tf = np.dot(A_tf,vp)
```
